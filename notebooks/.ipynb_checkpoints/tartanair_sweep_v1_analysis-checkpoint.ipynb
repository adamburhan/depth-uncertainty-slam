# Experiment description
## Goal
Test whether adding a depth-landscape-derived uncertainty metric to the existing octave-based keypoint uncertainty in ORB-SLAM improves SLAM trajectory estimation — specifically, whether it improves tracking robustness on fragile sequences without degrading accuracy on sequences that already track well.
## Setting
### Infrastructure
Using pyslam with the following tweaks:
- Containerized into a Singularity image and launched on the Mila cluster.
- `USE_CPP_CORE = False`: all runs use the Python core (C++ backend compiled but not active). Relative comparisons across presets are valid, but absolute numbers may differ from C++ runs.
- Deterministic ORB extraction enabled (`ORBextractorDeterministic`), cuDNN determinism flags on.
- Tracking-lost counter includes both LOST and RELOCALIZE frames → `percent_lost` is slightly broader than upstream definition.
### Dataset
TartanAir-v2, "easy" sequences only, 18 environments, 640×480 resolution, synthetic ground truth depth maps.
### Method
The depth-weighting function `_compute_kps_depth_weight` in `frame.py`:
1. Loads an auxiliary depth map (ground truth or Depth Anything V2 estimate).
2. Normalizes sentinels to NaN (inf, zero, negative values).
3. Computes **log-depth** gradient: $\sigma^2_d = \|\nabla \ln D\|^2$ via Sobel (ksize=3, scale=1/8 for true derivative units). Log-depth is scale-invariant: a 1m edge at 2m depth produces a large gradient, a 1m slope at 50m depth does not, eliminating the need for an arbitrary depth cap.
4. Pixels whose Sobel neighborhood touches any NaN are masked (gradient set to 0) to prevent gradients from NaN-fill.
5. Samples $\sigma^2_d$ at keypoint locations.
6. Computes per-keypoint weight $w = 1/(1 + \lambda \cdot \sigma^2_d)$.
7. **Median-1 normalizes** over valid-depth keypoints → robust to high-gradient outliers (mean normalization was sensitive to them). This redistributes rather than globally loosens influence, avoiding a confound with the chi²-threshold outlier rejection.
8. Keypoints with invalid depth or NaN-contaminated neighborhoods fall back to weight 1.0.

Weights are applied by multiplying `invSigma2` at four optimizer call sites (BA, pose optimization, local BA, LBA process).

#### Note on earlier version
An earlier sweep used raw depth gradients ($\|\nabla D\|^2$, ksize=5, no scale normalization) with a 20m depth cap and mean-1 normalization. The 20m cap replaced far pixels with 0, creating phantom Sobel edges at every sky/distance boundary. This caused systematic regressions on outdoor scenes. The log-depth formulation fixes this at the root.
### Presets
- `baseline`: octave-based keypoint uncertainty only.
- `depth_gt_{1,2,3}`: baseline × depth weight with $\lambda \in \{0.01, 0.1, 1.0\}$, $D$ = ground truth depth.
- `depth_est_{1,2,3}`: same, but $D$ = Depth Anything V2 estimate.
- Each (environment, preset, sequence) combination is run 10 times.
## Questions to answer
1. **Tracking robustness**: Does the method improve the fraction of seeds that successfully track (≤20% frames lost), paired by (env, sequence)? Measured via Wilcoxon signed-rank on per-sequence success rate differences.
2. **Conditional accuracy**: Among sequences where both baseline and method track reliably (≥7/10 seeds), does the method improve ATE? Measured via per-sequence median ATE, paired Wilcoxon.
3. **Do-no-harm**: On sequences where baseline already succeeds 10/10 seeds, does the method degrade accuracy? (Separate from Q2 — this isolates the "easy sequence" regime.)
4. **λ sensitivity**: How does the effect vary across λ ∈ {0.01, 0.1, 1.0}? Is the result robust to this choice or dependent on tuning?
5. **GT vs estimated depth**: Does the signal hold when using Depth Anything V2 predictions instead of ground truth? (Tests real-world applicability.)
6. **Characterization**: What distinguishes sequences where the method helps from those where it hurts? (Informs the anisotropic formulation.)

```python
import pandas as pd
```

```python
df = pd.read_parquet("/Users/adam/Documents/MILA/projects/depth-uncertainty-slam/results/tartanair_sweep_v1.parquet")
```

```python
df_ok = df[df.status == "ok"]
df_ok = df_ok[df_ok.percent_lost <= 20]
print(f"Completed: {len(df_ok)} / {len(df)}")
print(f"\nTracking loss <= 20%: {(df_ok.percent_lost <= 20).mean():.1%}")
print(f"\nBy preset:")
print(df_ok.groupby("preset").agg(
    n=("ate_rmse", "count"),
    median_ate=("ate_rmse", "median"),
    mean_pct_lost=("percent_lost", "mean"),
    frac_tracked=("percent_lost", lambda x: (x <= 20).mean())
).round(3).to_string())
```

```python
GT_PRESETS = ["baseline", "depth_gt_1", "depth_gt_2", "depth_gt_3"]

counts = df[df.status == "ok"].groupby(["env", "sequence", "preset"]).size().unstack("preset")
complete_gt = counts[counts[GT_PRESETS].eq(10).all(axis=1)].index

df_gt = (df[(df.status == "ok") & df.preset.isin(GT_PRESETS)]
         .set_index(["env", "sequence"]).loc[complete_gt].reset_index())

summary = df_gt.groupby(["env", "sequence", "preset"]).apply(
    lambda g: pd.Series({
        "success_rate": (g.percent_lost <= 20).mean(),
        "ate_median": g.loc[g.percent_lost <= 20, "ate_rmse"].median(),
        "ate_std": g.loc[g.percent_lost <= 20, "ate_rmse"].std(),
        "n_tracked": int((g.percent_lost <= 20).sum()),
    })
).reset_index()

pivot_mean = summary.pivot_table(index=["env", "sequence"], columns="preset", values="ate_median")
pivot_std = summary.pivot_table(index=["env", "sequence"], columns="preset", values="ate_std")
pivot_n = summary.pivot_table(index=["env", "sequence"], columns="preset", values="n_tracked")

# Display with n_tracked so you can see how many seeds actually tracked
pivot_display = pivot_mean.round(3).astype(str) + " ± " + pivot_std.round(3).astype(str) + " (n=" + pivot_n.astype(int).astype(str) + ")"
pivot_display[pivot_mean.isna()] = "—"

valid = pivot_mean.dropna(how="all", subset=GT_PRESETS).index
pivot_display = pivot_display.loc[valid]
pivot_display["best"] = pivot_mean.loc[valid, GT_PRESETS].idxmin(axis=1)

print(f"Sequences with all GT presets complete (10 seeds each): {len(pivot_display)}")
print(pivot_display.to_string())
print(f"\nBest preset counts:\n{pivot_display['best'].value_counts().to_string()}")
```

```python
from scipy.stats import wilcoxon
import numpy as np

GT_PRESETS = ["baseline", "depth_gt_1", "depth_gt_2", "depth_gt_3"]

# Restrict to sequences with all 4 GT presets having 10 seeds
counts = df[df.status == "ok"].groupby(["env", "sequence", "preset"]).size().unstack("preset")
complete = counts[counts[GT_PRESETS].eq(10).all(axis=1)].index
df_c = df[(df.status == "ok") & df.preset.isin(GT_PRESETS)].set_index(["env", "sequence"]).loc[complete].reset_index()

# --- Q1: Success rate ---
success = df_c.groupby(["env", "sequence", "preset"]).apply(
    lambda g: (g.percent_lost <= 20).mean()
).unstack("preset")

print("=== Q1: TRACKING SUCCESS RATE ===")
print(f"Mean success rate per preset:")
print(success[GT_PRESETS].mean().round(3).to_string())
print()

for p in ["depth_gt_1", "depth_gt_2", "depth_gt_3"]:
    diff = success[p] - success["baseline"]
    better = (diff > 0).sum()
    same = (diff == 0).sum()
    worse = (diff < 0).sum()
    stat, pval = wilcoxon(diff[diff != 0]) if (diff != 0).sum() >= 10 else (np.nan, np.nan)
    print(f"{p}: +{better} ={same} -{worse}, mean diff {diff.mean():+.3f}, Wilcoxon p={pval:.4f}")

# --- Q2: Conditional ATE (median per sequence, tracked seeds only) ---
ate = df_c[df_c.percent_lost <= 20].groupby(
    ["env", "sequence", "preset"]
).ate_rmse.median().unstack("preset")

print("\n=== Q2: CONDITIONAL ATE (median over tracked seeds) ===")
print(f"Grand median ATE per preset:")
print(ate[GT_PRESETS].median().round(4).to_string())
print()

for p in ["depth_gt_1", "depth_gt_2", "depth_gt_3"]:
    paired = ate[["baseline", p]].dropna()
    diff = paired[p] - paired["baseline"]
    better = (diff < 0).sum()
    worse = (diff > 0).sum()
    stat, pval = wilcoxon(diff) if len(diff) >= 10 else (np.nan, np.nan)
    print(f"{p}: ATE lower on {better}/{len(paired)} seqs, Wilcoxon p={pval:.4f}, median diff {diff.median():+.4f}")

# --- Q3: Combined score (for Liam summary) ---
print("\n=== Q3: HEADLINE NUMBERS ===")
for p in ["depth_gt_1", "depth_gt_2", "depth_gt_3"]:
    # Success rate improvement
    sr_diff = success[p] - success["baseline"]
    # ATE improvement (negative = better)
    paired = ate[["baseline", p]].dropna()
    ate_diff = paired[p] - paired["baseline"]
    print(f"\n{p} (λ={'0.01' if '1' in p[-1:] else '0.1' if '2' in p[-1:] else '1.0'}):")
    print(f"  Success rate: {sr_diff.mean():+.1%} avg change, p={wilcoxon(sr_diff[sr_diff!=0])[1]:.4f}")
    print(f"  ATE (when tracked): median change {ate_diff.median():+.4f}m, p={wilcoxon(ate_diff)[1]:.4f}")

# --- Q4: Characterize wins/losses ---
print("\n=== Q4: BIGGEST WINS AND LOSSES (depth_gt_2, λ=0.1) ===")
paired = ate[["baseline", "depth_gt_2"]].dropna()
diff = paired["depth_gt_2"] - paired["baseline"]
diff_sorted = diff.sort_values()

print("\nTop 10 wins (method improves ATE):")
for idx, val in diff_sorted.head(10).items():
    b, m = paired.loc[idx, "baseline"], paired.loc[idx, "depth_gt_2"]
    sr_b, sr_m = success.loc[idx, "baseline"], success.loc[idx, "depth_gt_2"]
    print(f"  {idx[0]:25s} {idx[1]}: {b:.3f} → {m:.3f} ({val:+.3f}m), success {sr_b:.0%}→{sr_m:.0%}")

print("\nTop 10 losses (method hurts ATE):")
for idx, val in diff_sorted.tail(10).items():
    b, m = paired.loc[idx, "baseline"], paired.loc[idx, "depth_gt_2"]
    sr_b, sr_m = success.loc[idx, "baseline"], success.loc[idx, "depth_gt_2"]
    print(f"  {idx[0]:25s} {idx[1]}: {b:.3f} → {m:.3f} ({val:+.3f}m), success {sr_b:.0%}→{sr_m:.0%}")
```

```python
MIN_TRACKED = 5

# Per (env, sequence, preset): median ATE and n_tracked
stats = df_c[df_c.percent_lost <= 20].groupby(
    ["env", "sequence", "preset"]
).agg(
    ate_median=("ate_rmse", "median"),
    n_tracked=("ate_rmse", "count")
).reset_index()

stats_piv_ate = stats.pivot_table(index=["env", "sequence"], columns="preset", values="ate_median")
stats_piv_n = stats.pivot_table(index=["env", "sequence"], columns="preset", values="n_tracked")

for p in ["depth_gt_1", "depth_gt_2", "depth_gt_3"]:
    # Both baseline and preset must have >= MIN_TRACKED seeds
    mask = (stats_piv_n["baseline"] >= MIN_TRACKED) & (stats_piv_n[p] >= MIN_TRACKED)
    paired = stats_piv_ate.loc[mask, ["baseline", p]].dropna()
    diff = paired[p] - paired["baseline"]
    
    better = (diff < 0).sum()
    worse = (diff > 0).sum()
    stat, pval = wilcoxon(diff) if len(diff) >= 10 else (np.nan, np.nan)
    
    print(f"\n{p} (n={len(paired)} sequences with >= {MIN_TRACKED} tracked seeds on both):")
    print(f"  ATE lower: {better}/{len(paired)}, higher: {worse}/{len(paired)}")
    print(f"  Median diff: {diff.median():+.4f}m, Wilcoxon p={pval:.4f}")
    print(f"  Mean diff: {diff.mean():+.4f}m")
```

```python
from scipy import stats
import matplotlib.pyplot as plt

# Your paired differences (method - baseline), filtered to stable sequences
mask = (stats_piv_n["baseline"] >= MIN_TRACKED) & (stats_piv_n[p] >= MIN_TRACKED)
paired = stats_piv_ate.loc[mask, ["baseline", "depth_gt_1"]].dropna()
diff = paired["depth_gt_1"] - paired["baseline"]

print(f"n = {len(diff)}")
print(f"mean = {diff.mean():.4f}, median = {diff.median():.4f}")
print(f"skewness = {stats.skew(diff):.2f}, kurtosis = {stats.kurtosis(diff):.2f}")
print(f"Shapiro-Wilk normality test: p = {stats.shapiro(diff)[1]:.4f}")

plt.hist(diff, bins=20)
plt.axvline(0, color='r')
plt.title("Distribution of paired ATE differences")
plt.savefig("paired_diff_hist.png")
```
